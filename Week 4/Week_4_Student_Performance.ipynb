```python
# Fungsi untuk plotting pie chart dengan warna estetik
def plot_pie_chart(column_name):

    data = df[column_name].value_counts()
    labels = data.index
    sizes = data.values

    # Warna estetik menggunakan palet "Pastel1" dari matplotlib
    colors = plt.get_cmap('Pastel1')(range(len(labels)))

    # Membuat sedikit efek explode pada setiap slice
    explode = [0.05] * len(labels)  # Membuat setiap slice terpisah sedikit

    plt.figure(figsize=(7,7))
    plt.pie(
        sizes,
        labels=labels,
        autopct='%1.1f%%',
        startangle=90,
        colors=colors,  # Menggunakan warna dari palet Pastel1
        explode=explode,  # Menambahkan efek explode
        shadow=True,  # Menambahkan efek bayangan
        wedgeprops={'edgecolor': 'black'}  # Menambahkan garis tepi hitam untuk memperjelas batas
    )
    plt.axis('equal')  # Untuk memastikan pie chart berbentuk lingkaran
    plt.title(f'Percentage of {column_name}', fontsize=14, color='darkblue')  # Judul dengan font lebih besar dan warna
    plt.show()

# Menampilkan pie chart untuk kolom bertipe kategorikal
plot_pie_chart('gender')
plot_pie_chart('race/ethnicity')
plot_pie_chart('lunch')
plot_pie_chart('parental level of education')
plot_pie_chart('test preparation course')
```

Prompt:

Beri kode python yang menunjukkan visualisasi nilai math, reading, dan writing siswa, dengan melihat pengaruh dari gender, race/ethnicity, parental level of education, lunch, test preparation course-nya!

```python
# Membuat kolom baru untuk kategori skor (low, medium, high) untuk masing-masing jenis skor
def categorize_score(score):
    if score < 60:
        return 'low'
    elif score < 80:
        return 'medium'
    else:
        return 'high'

df['math_category'] = df['math score'].apply(categorize_score)
df['reading_category'] = df['reading score'].apply(categorize_score)
df['writing_category'] = df['writing score'].apply(categorize_score)

# List kolom kategorikal dan kategori skor untuk looping
categorical_columns = ['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course']
score_categories = ['math_category', 'reading_category', 'writing_category']

# Membuat pie chart untuk setiap kombinasi kategori skor dan kolom kategorikal
for score_cat in score_categories:
    for col in categorical_columns:
        plt.figure(figsize=(6, 6))
        data = df.groupby([col, score_cat]).size().unstack().fillna(0)

        # Plot pie chart untuk setiap nilai unik dalam kolom kategorikal
        data.plot(kind='pie', subplots=True, autopct='%1.1f%%', startangle=90, legend=False, colormap='autumn')
        plt.suptitle(f'Percentage of {score_cat.replace("_category", "").capitalize()} by {col}')
        plt.ylabel('')
        plt.show()
```

Prompt:

Beri kode python untuk melihat jumlah siswa yang fail dan pass untuk masing-masing pelajaran (math, reading, writing score). Dengan minimal nilai yang dianggap pass adalah 75!



```python
# Menentukan ambang batas nilai untuk lulus
passing_score = 75

# Fungsi untuk menghitung jumlah siswa yang lulus dan gagal
def count_pass_fail(column):
    pass_count = (df[column] >= passing_score).sum()
    fail_count = (df[column] < passing_score).sum()
    return pass_count, fail_count

# Mencetak hasil untuk setiap pelajaran
subjects = ['math score', 'reading score', 'writing score']
results = {subject: count_pass_fail(subject) for subject in subjects}

# Menyiapkan data untuk visualisasi
labels = ['Pass', 'Fail']
data = {subject: list(result) for subject, result in results.items()}

# Membuat DataFrame untuk visualisasi
visual_data = pd.DataFrame(data, index=labels)

# Visualisasi dengan diagram batang
plt.figure(figsize=(10, 6))
visual_data.plot(kind='bar', color=['lightgreen', 'salmon'], edgecolor='black')
plt.title('Jumlah Siswa yang Lulus dan Gagal per Pelajaran', fontsize=16)
plt.xlabel('Status', fontsize=14)
plt.ylabel('Jumlah Siswa', fontsize=14)
plt.xticks(rotation=0)
plt.legend(title='Pelajaran', labels=['Math', 'Reading', 'Writing'])
plt.tight_layout()
plt.show()
```

Prompt: Beri kode untuk melihat nilai rata-rata seluruh mata pelajaran (math, reading writing) dari setiap siswa dan lihat korelasinya dengan masing-masing kategori

```python
# Menghitung nilai rata-rata untuk setiap siswa
df['average score'] = df[['math score', 'reading score', 'writing score']].mean(axis=1)

# Mengatur gaya seaborn
sns.set(style="whitegrid")

# Visualisasi rata-rata skor berdasarkan kategori
plt.figure(figsize=(20, 15))

# Rata-rata skor berdasarkan Gender
plt.subplot(2, 2, 1)
sns.boxplot(x='gender', y='average score', data=df, palette='pastel')
plt.title('Average Score by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Score')

# Rata-rata skor berdasarkan Race/Ethnicity
plt.subplot(2, 2, 2)
sns.boxplot(x='race/ethnicity', y='average score', data=df, palette='pastel')
plt.title('Average Score by Race/Ethnicity')
plt.xlabel('Race/Ethnicity')
plt.ylabel('Average Score')

# Rata-rata skor berdasarkan Parental Level of Education
plt.subplot(2, 2, 3)
sns.boxplot(x='parental level of education', y='average score', data=df, palette='pastel')
plt.title('Average Score by Parental Level of Education')
plt.xlabel('Parental Level of Education')
plt.ylabel('Average Score')
plt.xticks(rotation=45)

# Rata-rata skor berdasarkan Lunch
plt.subplot(2, 2, 4)
sns.boxplot(x='lunch', y='average score', data=df, palette='pastel')
plt.title('Average Score by Lunch Type')
plt.xlabel('Lunch Type')
plt.ylabel('Average Score')

plt.tight_layout()
plt.show()

# Menghitung korelasi antara kategori dan nilai rata-rata
correlation_matrix = df[['average score', 'math score', 'reading score', 'writing score']].corr()

# Visualisasi korelasi
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, linewidths=0.5)
plt.title('Correlation Matrix of Average Score and Subject Scores')
plt.show()
```
