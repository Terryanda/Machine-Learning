Polynomial Regression

```python
# Import library yang diperlukan

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
# Memuat data dari file CSV yang telah diunggah

data = pd.read_csv('/content/Position_Salaries.csv')
```

```python
# Melihat 5 data teratas untuk memastikan data terbaca dengan benar

data.head()
```

```python
# Memisahkan variabel independen (X) dan variabel dependen (y)

X = data[['Level']].values  # Mengambil kolom Level sebagai variabel independen
y = data['Salary'].values   # Mengambil kolom Salary sebagai variabel dependen
```

```python
# Membuat model regresi linear sebagai perbandingan awal

linear_regressor = LinearRegression()
linear_regressor.fit(X, y)
```

```python
# Membuat model regresi polinomial dengan derajat 4 (degree 4)

polynomial_features = PolynomialFeatures(degree=4)
X_poly = polynomial_features.fit_transform(X)
```

```python
# Membuat model regresi linear berdasarkan fitur polinomial

polynomial_regressor = LinearRegression()
polynomial_regressor.fit(X_poly, y)
```

```python
# Visualisasi hasil regresi linear dan polinomial

plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='red', label='Data Aktual')

# Prediksi dengan model regresi linear

plt.plot(X, linear_regressor.predict(X), color='blue', label='Regresi Linear')

# Prediksi dengan model regresi polinomial

X_grid = np.arange(min(X), max(X), 0.1)         # Membuat grid untuk visualisasi yang lebih halus
X_grid = X_grid.reshape((len(X_grid), 1))
plt.plot(X_grid, polynomial_regressor.predict(polynomial_features.transform(X_grid)), color='green', label='Regresi Polinomial (Degree 4)')

# Menambahkan judul dan label
plt.title('Perbandingan Regresi Linear dan Polinomial')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.legend()
plt.grid(True)
plt.show()
```
